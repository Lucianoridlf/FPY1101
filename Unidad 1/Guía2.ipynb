Una empresa de tecnología quiere optimizar el tiempo de carga y descarga de archivos en su servicio de almacenamiento en la nube. Los tiempos medios para diferentes tamaños de archivos son los siguientes:

<p><center>

|-------------------------|---------------------|-----------------------|

|$$250$$                  |$$50$$               |$$40$$                 |
|$$500$$                  |$$100$$              |$$80$$                 |
|$$1.000$$                 |$$200$$              |$$160$$                |

</center></p>

1. Encuentra las funciones que modelan el tiempo de carga $f(x)$ y el tiempo de descarga $g(x)$ en función del tamaño del archivo $x$. Utiliza la función *polyfit* de *Numpy* (debes encontrar dos funciones).

3. Calcula el tiempo de carga y descarga para un archivo de $750$ MB.

# 3.1
"funcion -> y = mx + b"

# 3.2 
Variable dependiente = tiempo carga y descarga (s)
variable independiente = tamaño archivo(mb)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

tamanio = np.array([50, 100, 250, 500, 1000])
carga = np.array([10, 20, 50, 100, 200])
descarga = np.array([8, 16, 40, 80 ,160])

#ajuste

mb_carga = np.polyfit(tamanio, carga, 1)
pol_carga = np.poly1d(mb_carga, 1)

mb_descarga = np.polyfit(tamanio, descarga, 1)
pol_descarga = np.poly1d(mb_descarga, 1)

print(f'La pendiente de la primera funcion de carga es: {mb_carga[0]}, el cof es: {mb_carga[1]}, el resultado es: {pol_carga}')
print(f'La pendiente de la segunda funcion de carga es: {mb_descarga[0]}, el cof es: {mb_descarga[1]}, el resultado es: {pol_descarga}')


#3.3
def tiempo_carga(x):
    return  0.2 * x

def tiempo_descarga(x):
    return 0.16 * x

print(f'El tiempo de carga para un archivo de 750 mb es {tiempo_carga(750)} segundos')
print(f'El tiempo de carga para un archivo de 750 mb es {tiempo_descarga(750)} segundos')
    
```

# 3.4

```python

def t_carga(x):
    return 0.2 * x - 163

def t_descarga(x):
    return 0.16 * x - 163

sol_carga = fsolve(t_carga, 56)
sol_descarga = fsolve(t_descarga, 56)
print (sol_carga)

print(f'El tamaño del archivo al tener un tiempo de carga de 163 segundos es: {sol_carga[0]:.0f} mb')

dist_3 = np.linspace(tamanio.min(), tamanio.max(), 100)
plt.plot(dist_3, t_carga(dist_3), label = " t. carga")
plt.plot(dist_3, t_descarga(dist_3), label = "t. descarga")
plt.grid(True)
plt.xlabel('Tamanio')
plt.title('carga versus descarga en el tiempo')
plt.legend(shadow= True, frameon= True)
plt.show()
```
