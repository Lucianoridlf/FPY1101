# Pregunta 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

#Funciones del enunciado
def V(t):
    return 0.85 * (t ** 2) + 13.5 * t + 38

def S(t):
    return 1.2 * (t **2) + 16.2 * t + 29

def C(t):
    return 1.4 * (t**2) + 18.7 * t + 27
```

```python
#Cantidad de lluvia Valparaiso (mm)

print(f'La cantidad de lluvia caida en Valparaiso al cabo de 55 min es: {V(55/60):.2f} mm')
```

```python
#Cantidad de lluvia Santiago (m)

print(f'La cantidad de lluva caida en Santiago al cabo de 55 min es: {S(55/60):.2f}')
```

```python
#Cantidad de lluvia Concepcion (m)

print(f'La cantidad de lluvia caida en Concepción al cabo de 55 min es: {C(55/60):.2f} mm')
```

```python
dist = np.linspace(0, 24, 100)

plt.plot(dist, V(dist))
plt.grid(True)
plt.title('Lluvia en Valparaiso')
plt.ylabel("cantidad de lluvia (mm)")
plt.xlabel("Tiempo (h)")
plt.legend()
plt.show()
```

```python
#Grafico para Santiago

dist = np.linspace(0, 24, 100)

plt.plot(dist, S(dist))
plt.grid(True)
plt.title('Lluvia en Santiago')
plt.ylabel("cantidad de lluvia (mm)")
plt.xlabel("Tiempo (h)")
plt.legend()
plt.show()
```

```python
#grafico para Concepcion

dist = np.linspace(0, 24, 100)

plt.plot(dist, C(dist))
plt.grid(True)
plt.title('Lluvia en Concepción')
plt.ylabel("cantidad de lluvia (mm)")
plt.xlabel("Tiempo (h)")
plt.legend()
plt.show()
```

# Pregunta 4

```python
#Valparaiso 

def V(t):
    return (0.85 * (t ** 2) + 13.5 * t + 38 ) - 95
dist = np.linspace(0, 24, 100)
valpo = fsolve(V, 4)
print(f'La cantidad de horas para que Valparaiso necesite de los equipos de emergencia es {valpo[0]:.2f} horas')
```

```python
#Santiago
def S(t):
    return (1.2 * (t **2) + 16.2 * t + 29) - 95
dist = np.linspace(0, 24, 100)
santiago = fsolve(S, 4)
print(f'La cantidad de horas para que Valparaiso necesite de los equipos de emergencia es {santiago[0]:.2f} horas')
```

```python
#Concepcion

def C(t):
    return (1.4 * (t**2) + 18.7 * t + 27) - 95
dist = np.linspace(0, 24, 100)
conce = fsolve(C, 4)
print(f'La cantidad de horas para que Valparaiso necesite de los equipos de emergencia es {conce[0]:.2f} horas')
```

# Pregunta 5

Los equipos de las primeras 6 horas no deben estar frecuentemente revisando los alcantarillados ya que estos no acumulan los 95 milimetros de lluvia hasta recien alcanzar las 3 horas aproximadamente en cada ciudad. Por lo tanto, los equipos de las últimas 6 horas deben estar con mayor frecuencia revisando los alcantarillados debido a que la acumulación de lluvia que se genera posterior a las 6 horas es mayor. Por ende, el segundo equipo de emergencia es el que va con mayor frecuencia. 

```python
dist = np.linspace(0, 12, 100)

plt.plot(dist, V(dist), label= "Valparaiso")
plt.plot(dist, S(dist), label= "Santiago")
plt.plot(dist, C(dist), label= "Concepcion")
plt.legend()
plt.grid(True)
plt.title('Equipos de emergencia por ciudad')
plt.ylabel("cantidad de lluvia (mm)")
plt.xlabel("Tiempo (h)")
plt.legend()
plt.show()
```
